```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
df_smto = yf.download("SMTO3.SA")
smto = df_smto["Adj Close"]
df_smto
```

```python
df_amer = yf.download("AMER3.SA")
amer = df_amer["Adj Close"]
df_amer
```

```python
df_itsa = yf.download("ITSA4.SA")
itsa = df_itsa["Adj Close"]
df_itsa
```

```python
df_ibov = yf.download("^BVSP", start = "2005-01-01", end = "2023-06-21")
ibov = df_ibov["Adj Close"]
df_ibov
```

```python
fechamento = [ibov, smto, amer, itsa]

df = pd.DataFrame(fechamento)
df = df.T
df = df.dropna()
df.columns = ["ibov", "smto", "amer", "itsa"]
df
```

```python
df['Retorno IBOV'] = df["ibov"].pct_change()
df['Retorno SMTO'] = df["smto"].pct_change()
df['Retorno AMER'] = df["amer"].pct_change()
df['Retorno ITSA'] = df["itsa"].pct_change()
df
```

```python
fig, ax = plt.subplots(figsize = (15, 10))

intervalo = [pd.Timestamp("2022-06-01"), pd.Timestamp("2023-06-01")]

ax.plot(df.index, df['Retorno IBOV'], label = "Ibovespa", color = "red")
ax.plot(df.index, df['Retorno SMTO'], label = "São Martinho", color = "green")
ax.plot(df.index, df['Retorno AMER'], label = "Americanas", color = "blue")
ax.plot(df.index, df['Retorno ITSA'], label = "Itaúsa", color = "yellow")

ax.legend(fontsize = 12) # Box de legenda

ax.set_title("Retorno diário", fontsize = 16) # Título do gráfico

ax.set_xlim(intervalo)
ax.set_ylim(-0.8, 0.3)
ax.set_xlabel("Período")
ax.set_ylabel("Retorno")

ax.grid()
```

```python
df['IBOV acumulado'] = (df['Retorno IBOV']+1).cumprod()
df['SMTO acumulado'] = (df['Retorno SMTO']+1).cumprod()
df['AMER acumulado'] = (df['Retorno AMER']+1).cumprod()
df['ITSA acumulado'] = (df['Retorno ITSA']+1).cumprod()
df
```

```python
fig, ax = plt.subplots(figsize = (15, 10))

intervalo = [pd.Timestamp("2022-06-01"), pd.Timestamp("2023-06-01")]

ax.plot(df.index, df['IBOV acumulado'], label = "Ibovespa", color = "red")
ax.plot(df.index, df['SMTO acumulado'], label = "São Martinho", color = "green")
ax.plot(df.index, df['AMER acumulado'], label = "Americanas", color = "blue")
ax.plot(df.index, df['ITSA acumulado'], label = "Itaúsa", color = "yellow")

ax.legend(fontsize = 12) # Box de legenda

ax.set_title("Retorno acumulado", fontsize = 16) # Título do gráfico

ax.set_xlabel("Período")
ax.set_ylabel("Retorno")

ax.grid()
```
