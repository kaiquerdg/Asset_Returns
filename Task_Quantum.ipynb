```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_itub = yf.download("ITUB4.SA", start = "2022-08-18", end = "2023-08-18")
itub = df_itub["Adj Close"]

df_bbdc = yf.download("BBDC4.SA", start = "2022-08-18", end = "2023-08-18")
bbdc = df_bbdc["Adj Close"]

df_b3sa = yf.download("B3SA3.SA", start = "2022-08-18", end = "2023-08-18")
b3sa = df_b3sa["Adj Close"]

df_sanb = yf.download("SANB4.SA", start = "2022-08-18", end = "2023-08-18")
sanb = df_sanb["Adj Close"]

df_bbas = yf.download("BBAS3.SA", start = "2022-08-18", end = "2023-08-18")
bbas = df_bbas["Adj Close"]
```

```python
matriz_correlacao = {'Itaú': itub, 'Bradesco': bbdc, 'B3': b3sa, 'Santander': sanb,
                     'Banco do Brasil': bbas}
df_correl = pd.DataFrame(matriz_correlacao)

fig, ax = plt.subplots(figsize = (18, 10))
correlacao = df_correl.corr()

sns.heatmap(correlacao, annot=True, cmap="coolwarm")
plt.title('Gráfico de Correlação')
```

```python
fechamento = [itub, bbdc, b3sa, sanb, bbas]
df = pd.DataFrame(fechamento)
df = df.T
df = df.dropna()
df.columns = ["ITUB4", "BBDC4", "B3SA3", "SANB4", "BBAS3"]
```

```python
df.iloc[:, 0:5].max()
```

```python
df.iloc[:, 0:5].min()
```

```python
df['Retorno ITUB4'] = df["ITUB4"].pct_change()
df['Retorno BBDC4'] = df["BBDC4"].pct_change()
df['Retorno B3SA3'] = df["B3SA3"].pct_change()
df['Retorno SANB4'] = df["SANB4"].pct_change()
df['Retorno BBAS3'] = df["BBAS3"].pct_change()
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
ax.plot(df.index, df['Retorno ITUB4'], label = "Itaú", color = "red")
ax.legend(fontsize = 12) # Box de legenda
ax.set_title("Retorno diário", fontsize = 16) # Título do gráfico
ax.set_xlabel("Período")
ax.set_ylabel("Retorno")
ax.grid()
```

```python
df['ITUB4 acumulado'] = (df['Retorno ITUB4']+1).cumprod()
df['BBDC4 acumulado'] = (df['Retorno BBDC4']+1).cumprod()
df['B3SA3 acumulado'] = (df['Retorno B3SA3']+1).cumprod()
df['SANB4 acumulado'] = (df['Retorno SANB4']+1).cumprod()
df['BBAS3 acumulado'] = (df['Retorno BBAS3']+1).cumprod()
```

```python
fig, ax = plt.subplots(figsize = (15, 10))

ax.plot(df.index, df['ITUB4 acumulado'], label = "Itaú", color = "red")
ax.plot(df.index, df['BBDC4 acumulado'], label = "Bradesco", color = "green")
ax.plot(df.index, df['B3SA3 acumulado'], label = "B3", color = "blue")
ax.plot(df.index, df['SANB4 acumulado'], label = "Santander", color = "orange")
ax.plot(df.index, df['BBAS3 acumulado'], label = "Banco do Brasil", color = "purple")

ax.legend(fontsize = 12) # Box de legenda

ax.set_title("Retorno acumulado", fontsize = 16) # Título do gráfico

ax.set_xlabel("Período")
ax.set_ylabel("Retorno")

ax.grid()
```
